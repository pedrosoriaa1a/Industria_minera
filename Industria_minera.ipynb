# Minas operativas de recursos minerales


El conjunto de datos de Minas y Recursos Minerales proporciona información completa sobre minas operativas de recursos minerales, excluidas las canteras de arena y grava. Incluye datos detallados sobre la ubicación, los datos de contacto y las características específicas de estas minas.

Un aspecto clave de este conjunto de datos es que se centra exclusivamente en minas en operación activa. Esto garantiza que la información esté actualizada y sea relevante para las diversas partes interesadas, incluidos los equipos de primeros auxilios y las fuerzas del orden.

Para facilitar el fácil acceso a estas minas, el punto de ubicación de cada mina se ha ubicado estratégicamente en la intersección entre el camino de transporte principal dentro del sitio de la mina y el camino público más cercano. Este punto sirve como punto de partida desde el cual debe ser sencillo navegar a las áreas mineras o pozos mineros reales.

El conjunto de datos cubre una amplia gama de atributos relacionados con cada mina. Estos incluyen descripciones detalladas de la clasificación de la industria utilizando códigos NAICS (Sistema de clasificación de la industria de América del Norte) y códigos SIC (clasificación industrial estándar). Además, la información sobre la clasificación de seguridad destaca cualquier consideración de seguridad potencial asociada con cada mina.

Se proporcionan datos de contacto para cada mina para garantizar una comunicación eficiente en caso de emergencias o consultas. Estos incluyen números de teléfono de contacto de emergencia junto con números de extensión, títulos de personas de contacto de emergencia, proveedores responsables de proporcionar servicios de datos relacionados con una mina en particular, el nombre del oficial de inspección y los números de teléfono generales para contactar a las respectivas compañías mineras o plantas.

La información de dirección comprende direcciones completas que incluyen detalles de dirección adicionales cuando corresponda, como números de suite o de piso. Los nombres de las ciudades se especifican junto con los condados donde se encuentran estas minas para identificar sus ubicaciones geográficas con mayor precisión. Los códigos postales proporcionan referencias de códigos postales esenciales, mientras que los códigos FIPS (Estándares Federales de Procesamiento de Información) ofrecen identificadores únicos para cada condado.

Las indicaciones desde las vías públicas cercanas guían a las personas hacia el acceso eficiente a minas específicas. Esto garantiza que los socorristas o los visitantes puedan llegar fácilmente a sus destinos previstos dentro de estas extensas áreas mineras.

El conjunto de datos también incluye datos geoespaciales valiosos como las coordenadas de latitud (y) y longitud (x), lo que permite un posicionamiento preciso de cada mina en un mapa. El nivel de precisión de estos puntos de datos geoespaciales se proporciona para comprender la exactitud y confiabilidad de esta información de ubicación.

La categorización temática basada en el Programa de Infraestructura de Seguridad Nacional ofrece información sobre el enfoque industrial específico o el propósito de cada mina. Esta clasificación permite una mejor comprensión y organización dentro del conjunto de datos.

Además, se proporciona información de control y garantía de calidad para transmitir la confiabilidad y validez de los datos. Esto incluye detalles sobre cómo se determinaron los datos geoespaciales (geohow) y las fechas registradas tanto para la información geoespacial (geodate) como para la información de contacto general (contdate).

### Conociendo la variable objetivo (target)

Basándonos en el contexto proporcionado sobre el conjunto de datos se pueden inferir los valores de la variable objetivo "stat_code" de la siguiente manera:

* Valor numérico 1: Indica que la mina está en estado operativo activo y plenamente funcional. Esta clasificación implica que la mina está en funcionamiento y no hay observaciones ni problemas significativos que afecten su operación.

* Valor numérico 2: Representa una mina en estado operativo con observaciones. Aunque la mina sigue estando en funcionamiento, pueden existir observaciones o preocupaciones menores que requieren atención, como pequeñas irregularidades o necesidades de mantenimiento.

* Valor numérico 3: Se asigna a minas que están en proceso de cierre. Estas minas pueden estar en el proceso de cierre gradual de sus operaciones o en la etapa final de cierre, con actividades de desmantelamiento y restauración en curso.

* Valores alfabéticos (A, B, C, F): Estos valores alfabéticos pueden indicar subcategorías dentro de los estados operativos numéricos mencionados anteriormente.
  * "A" representa una mina en estado activo con observaciones menores,
  * "B" representa una mina en proceso de cierre con pendientes regulatorias,
  * "C" Se asigna a minas inactivas con condiciones especiales
  * "F" representa minas cerradas definitivamente.

* Valor vacío: Indica que no hay información disponible sobre el estado operativo de la mina





Debido a la necesidad de clasificar el estado operativo de las minas en función de diversos factores relevantes para la industria minera. Aquí están algunas razones que respaldan esta elección:

* Relevancia para las partes interesadas: El estado operativo de una mina es una información crucial para diversas partes interesadas, incluidos los equipos de primeros auxilios, las fuerzas del orden, las autoridades regulatorias y las propias empresas mineras. Comprender el estado operativo de una mina puede ayudar a garantizar una respuesta rápida y efectiva en caso de emergencia, así como a facilitar la supervisión y el cumplimiento normativo.

* Actualización y relevancia de los datos: Dado que el conjunto de datos se centra exclusivamente en minas en operación activa, la información proporcionada sobre el estado operativo de cada mina debe estar actualizada y ser relevante. Esto asegura que los modelos de machine learning entrenados utilizando esta variable objetivo estén basados en datos actualizados y reflejen fielmente la realidad de las operaciones mineras en curso.

* Facilita la comunicación y la coordinación: La variable objetivo "stat_code" proporciona una forma estandarizada de comunicar el estado operativo de las minas, lo que facilita la coordinación entre diferentes partes interesadas y promueve una mejor gestión y supervisión de las operaciones mineras.

### Encoding

Para obtener las 'features' que se van a utilizar para obtener una predicción del modelo es necesario codificar algunas de ellas por el tipo de dato que son.



````python
# Funcion que implementa Label Encoding a cualquier variable dentro del dataset
def label_encode_variable(dataset, variable_name):
    """
    Aplica Label Encoding a una variable específica en el dataset.

    Parámetros:
    - dataset: DataFrame de pandas que contiene el dataset.
    - variable_name: Nombre de la variable a la que se aplicará Label Encoding.

    Retorna:
    - El dataset con la variable label encoded añadida.
    """
    le = LabelEncoder()
    dataset[variable_name + '_label_encoded'] = le.fit_transform(dataset[variable_name])
    return dataset
````

````python
# Columnas tipo object
columnas_object = min_data.select_dtypes(include=['object']).columns
# Aplicando función label_encode_variable
for columna in columnas_object:
    min_data = label_encode_variable(min_data, columna)
````

## Elección del modelo

### Random Forest

#### Dividir dataset

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20,
                                                        random_state = 1998)
````

#### Ejecutando modelo

````python
# Creamos el modelo
tree = RandomForestClassifier(n_estimators = 100, class_weight='balanced')

# Ajustamos el modelo a los datos de entrenamiento
tree.fit(X_train, y_train)

# Realizamos la predicción
y_pred = tree.predict(X_test)
````

#### Evaluando modelo

````python
# Score en datos de entrenamiento
tree.score(X_train, y_train)
````

````python
# Score en datos de test
tree_accuracy = tree.score(X_test, y_test)
tree_accuracy
````

````python
# Matriz de confusión
cm = confusion_matrix(y_test, y_pred, labels=tree.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                             display_labels=tree.classes_)
disp.plot(cmap=plt.cm.Blues)
plt.show()
````

````python
print(classification_report(y_test, y_pred))
````

````python
param_range = np.arange(1, 101, 10)

train_scores, test_scores = validation_curve(
                                              tree,
                                              X, y,
                                              param_name="n_estimators",
                                              param_range=param_range,
                                              cv=5,
                                              scoring="accuracy",
                                              n_jobs=-1
                                          )

# Calcular la media y desviación estándar de los puntajes de entrenamiento y prueba
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)

# Grafica
plt.figure(figsize=(10, 6))
plt.plot(param_range, train_mean, label="Training score", color="blue")
plt.plot(param_range, test_mean, label="Cross-validation score", color="green")
plt.fill_between(param_range, train_mean - train_std, train_mean + train_std, color="blue", alpha=0.2)
plt.fill_between(param_range, test_mean - test_std, test_mean + test_std, color="green", alpha=0.2)
plt.title("Curvas de validación de RandomForestClassifier")
plt.xlabel("Número de estimadores")
plt.ylabel("Accuracy Score")
plt.xticks(param_range)
plt.legend(loc="best")
plt.show()
````

### KNN

#### Dividir dataset

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20,
                                                        random_state = 1998)
````

#### Ejecutando modelo

````python
# Crear el modelo KNN
knn = KNeighborsClassifier(n_neighbors = 5)

# Entrenar el modelo
knn.fit(X_train, y_train)

# Realizar predicciones
y_pred = knn.predict(X_test)
````

#### Evaluando modelo

````python
# Matriz de confusión
cm = confusion_matrix(y_test, y_pred, labels=knn.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                             display_labels=knn.classes_)
disp.plot(cmap=plt.cm.Blues)
plt.show()
````

````python
print(classification_report(y_test, y_pred))
````

````python
knn_accuracy = knn.score(X_test, y_test)
knn_accuracy
````

````python
# Curvas de validación

param_range = np.arange(1, 11)

train_scores, test_scores = validation_curve(
                                              knn,
                                              X, y,
                                              param_name="n_neighbors",
                                              param_range=param_range,
                                              cv=5,
                                              scoring="accuracy",
                                              n_jobs=-1
                                            )

# Calcular la media y desviación estándar de los puntajes de entrenamiento y prueba
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)

# Graficar las curvas de validación
plt.figure(figsize=(10, 6))
plt.plot(param_range, train_mean, label="Training score", color="blue")
plt.plot(param_range, test_mean, label="Cross-validation score", color="green")
plt.fill_between(param_range, train_mean - train_std, train_mean + train_std, color="blue", alpha=0.2)
plt.fill_between(param_range, test_mean - test_std, test_mean + test_std, color="green", alpha=0.2)
plt.title("Curva de validación de KNN")
plt.xlabel("Número de vecinos")
plt.ylabel("Accuracy Score")
plt.xticks(param_range)
plt.legend(loc="best")
plt.show()
````

#### Conclusiones del modelo

Se evaluó el rendimiento del modelo KNN variando el número de vecinos (desde 1 hasta 10) y observando cómo afecta esto a la precisión del modelo en el conjunto de entrenamiento y en el conjunto de validación cruzada.

* Curva de entrenamiento:
Cuando el número de vecinos es 2, la precisión en el conjunto de entrenamiento es de 0.8, lo que significa que el modelo clasifica correctamente el 80% de las muestras del conjunto de entrenamiento cuando se utiliza un vecino cercano para la clasificación.
Cuando el número de vecinos aumenta a 10, la precisión en el conjunto de entrenamiento disminuye a 0.72, lo que indica que el modelo clasifica correctamente el 72% de las muestras del conjunto de entrenamiento cuando se consideran más vecinos para la clasificación. Esto sugiere que agregar más vecinos puede introducir más ruido en la clasificación y reducir la precisión en el conjunto de entrenamiento.

* Curva de validación cruzada:
Para 2 vecinos, la precisión en el conjunto de validación cruzada es de 0.55, lo que indica que el modelo clasifica correctamente el 55% de las muestras del conjunto de validación cruzada. Esto sugiere que el modelo puede estar sobreajustando los datos en esta configuración, ya que la precisión en el conjunto de validación cruzada es significativamente menor que en el conjunto de entrenamiento.
Sin embargo, cuando el número de vecinos aumenta a 10, la precisión en el conjunto de validación cruzada aumenta ligeramente a 0.58. Esto sugiere que el modelo puede generalizar mejor con más vecinos en esta configuración, ya que la precisión en el conjunto de validación cruzada es más cercana a la precisión en el conjunto de entrenamiento.

En resumen, un número menor de vecinos puede llevar a un modelo con alta varianza (sobreajuste), mientras que un número mayor de vecinos puede llevar a un modelo con alto sesgo (subajuste). En este caso, podríamos considerar un número intermedio de vecinos, como 5, que puede proporcionar un equilibrio entre el sesgo y la varianza y producir un mejor rendimiento general en el conjunto de validación cruzada.
